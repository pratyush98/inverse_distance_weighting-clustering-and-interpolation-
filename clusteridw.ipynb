```python
latu=[]
longu=[]
elevu=[]
for i in range(1,len(c),1):
    latu.append(float(c[i][1]))
    longu.append(float(c[i][2]))
    elevu.append(float(c[i][3]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
from sklearn.cluster import KMeans
plt.scatter(latu,longu)
plt.show()
```

```python
cat=np.c_[latu,longu]
kmeans = KMeans(n_clusters=13)
kmeans.fit(cat)

centroids = kmeans.cluster_centers_
labels = kmeans.labels_

print(centroids)
print(labels)
```

```python
import math 
import numpy as np
def clusterinterp(latu,longu,elevu,non=0):
    p=1
    lat=[]
    longi=[]
    elev=[]
    lat=latu
    longi=longu
    elev=elevu
    diff=1
    XX=np.sort(lat)
    YY=np.sort(longi)
    xmin=XX[0]
    xmax=XX[len(lat)-1]
    ymin=YY[0]
    ymax=YY[len(longi)-1]
    rows=(xmax-xmin)/diff
    col=(ymax-ymin)/diff
    col=int(col)
    rows=int(rows)
    data=np.zeros((rows,col))
    for i in range(len(lat)):
            if int(lat[i]/diff-xmin/diff)<rows and int(longi[i]/diff-ymin/diff)<col:
                      data[int(lat[i]/diff-xmin/diff)][int(longi[i]/diff-ymin/diff)]=elev[i]
    inc=0.005/len(latu)*(rows*col)
    sump=0
    sumwie=0
    non=non+len(lat)
    inc=int(inc)
    print(inc)
    for i in range(0,data.shape[0],10):
        for j in range(0,data.shape[1],10):
            if data[i][j]==0:
                for k in range(len(lat)):
                    div=math.pow(np.sqrt(((lat[k]-xmin)/diff-i)*((lat[k]-xmin)/diff-i)+((longi[k]-ymin)/diff-j)*((longi[k]-ymin)/diff-j)),p)
                    if int(lat[k]/diff-xmin/diff)<rows and int(longi[k]/diff-ymin/diff)<col:
                        sumwie=sumwie+data[int(lat[k]/diff-xmin/diff)][int(longi[k]/diff-ymin/diff)]/div
                        sump=sump+1/div
                
                data[i][j]=sumwie/sump
                latout.append(i*diff+xmin)
                longout.append(j*diff+ymin)
                elevout.append(data[i][j])
                non=non+1
            sumwie=0
            sump=0
    print(non)
    print(np.count_nonzero(data))
```

```python
def super(no):
    sendlat=[]
    sendlong=[]
    sendelev=[]
    for i in range(len(latu)):
        if labels[i]==no:
            sendlat.append(latu[i])
            sendlong.append(longu[i])
            sendelev.append(elevu[i])       
    clusterinterp(sendlat,sendlong,sendelev)  
```
